### Categorical Variable

```python
df.generation_id.value_counts()
```

```python
# import the packages needed
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df
```

```python
sns.color_palette()[3]
```

```python
sns.countplot(data=df,x='generation_id', color = 'red');
plt.title('Specie Count By Generation')
plt.xlabel('Generation ID')
plt.ylabel('Frequency');
```

```python
df.generation_id.value_counts()
```

```python
x= df.generation_id.value_counts().index
y = df.generation_id.value_counts().values
```

```python
plt.bar(x= x, height= y);
```

```python
plt.pie(df.generation_id.value_counts(),
        labels=df.generation_id.value_counts().index,
       wedgeprops = {'width': 0.5});
```

```python
import numpy as np
```

```python
df.weight.mean()
```

```python
np.mean(df.weight)
```

```python
df['size'] = np.where(df.weight > np.mean(df.weight), 'Overweight', 'Underweight')
```

```python
df
```

```python
sns.countplot(data=df, x='generation_id', hue='size');
```
